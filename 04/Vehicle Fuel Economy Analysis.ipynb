```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# fuel = pd.read_csv('https://www.kaggle.com/epa/fuel-economy#database.csv')

fuel = pd.read_csv('database.csv', header=0, low_memory=False)
```

```python
fuel.corr()
```

```python
sns.heatmap(fuel.corr())
```

```python
fuel.sum()
```

```python
fueleff = fuel[['Year', 'Make', 'City MPG (FT1)', 'City MPG (FT2)', 'Highway MPG (FT1)', 'Highway MPG (FT2)', 'Combined MPG (FT1)', 'Combined MPG (FT2)']]
```

```python
names = ['yr', 'make', 'cmpg1', 'cmpg2', 'hmpg1', 'hmpg2', 'compg1', 'compg2']
fueleff.columns = names
```

```python
graphdata = fueleff.groupby(['yr','make'])[['compg1']].mean()
```

```python
graphdata.plot(figsize = [16,8]);
```
